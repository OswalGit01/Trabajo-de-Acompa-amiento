```python
import pandas as pd
import matplotlib as plt
import seaborn as sns

df1= pd.read_csv("diabetes.csv")

print(df1.columns)
df1
```

```python
df1.info()
```

```python
import matplotlib.pyplot as plt

plt.title('Histograma Glucosa')
plt.hist(df1.Glucose, color='darkred', alpha=0.8, label='Glucosa')
plt.ylabel('Frecuencia', fontsize=10)
plt.xlabel('Glucosa', fontsize=10)
plt.grid(True)
plt.show()
```

```python
plt.title('Histograma Insulina')
plt.hist(df1.Insulin,color='darkred',alpha=0.8  )
plt.ylabel('Frecuencia', fontsize=10)
plt.xlabel('Insulina', fontsize=10)
plt.grid(True)
plt.show()
```

```python
plt.title('Histogram BloodPressure')
plt.hist(df1.BloodPressure,color='darkred',alpha=0.8)
plt.ylabel('Frecuencia', fontsize=10)
plt.xlabel('PresiÃ³n Sanguinea', fontsize=10)
plt.grid()
```

```python
df1["Pregnancies"].value_counts().plot(
    kind="bar",
    title="Pregnancies",
    color="darkred"
)

plt.xlabel("Numero de embarazos")
plt.ylabel("Cuenta")

plt.show()
```

```python
df1["Outcome"].value_counts().plot(
    kind="bar",
    title="Diabetes",
    color='darkred'
)

plt.xlabel("Â¿Tiene diabetes?")
plt.ylabel("Cuenta")

plt.show()
```

```python
df1["Outcome"].value_counts().plot(
    kind="pie",
    colors=["darkred", "darkorange"],
    title='Distribucion de la variable de salida',
    labels=["No Diabetes", "Diabetes"],
    ylabel=''
)
```

```python
df1.describe()
```

```python
fig = plt.figure(figsize=(14,8))
sns.boxplot(data=df1)
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(
    data=df1,
    x="Glucose",
    y="Age",
    hue="Outcome",
    palette={0: "darkred", 1: "darkorange"},
    alpha=0.7,
    s=60
)

plt.title("Scatter Plot de glucosa vs edad por variable salida")
plt.xlabel("Glucosa")
plt.ylabel("Edad")

plt.legend(title="Variable salida", labels=["No Diabetes", "Diabetes"])

plt.show()
```
